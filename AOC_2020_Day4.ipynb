```javascript
const formatPassportInput = (stringyInput) => {
    const splitByEmptyLine = stringyInput.split(/\n\s*\n/);
    const passportArray = splitByEmptyLine.map((entry) => {
        return entry.split(/[ \n]/)
    });
    const passportObjectArray = passportArray.map((passport) => {
        return passport.reduce((acc, curr) => {
            const [key, val] = curr.split(":");
            return {...acc, [key]: val}
        }, {})
    })
    return passportObjectArray
}
```

```javascript
const byrRegex = RegExp('(?:^19[2-9][0-9]$)|(?:^200[0-2]$)');
const iyrRegex = RegExp('(?:^201[0-9]$)|(?:^2020$)')
const eyrRegex = RegExp('(?:^202[0-9]$)|(?:^2030$)');
const hgtRegex = RegExp('(?:^1[5-8][0-9]cm$)|(?:^19[0-3]cm$)|(?:^59in$)|(?:^6[0-9]in$)|(?:^7[0-6]in$)');
const hclRegex = RegExp('^#[0-9a-f]{6}$');
const eclRegex = RegExp('^amb$|^blu$|^brn$|^gry$|^grn$|^hzl$|^oth$');
const pidRegex = RegExp('^[0-9]{9}$');
```

```javascript
class ValidPart2PassportInterface {
    constructor(passport) {
        this.byr = passport.byr;
        this.iyr = passport.iyr;
        this.eyr = passport.eyr;
        this.hgt = passport.hgt;
        this.hcl = passport.hcl;
        this.ecl = passport.ecl;
        this.pid = passport.pid;
        this.cid = passport.cid;
        this.valid = 1;
        // rules for part 1
        if (!this.byr || !this.iyr || !this.eyr || !this.hgt || !this.hcl || !this.ecl || !this.pid) {
            this.valid = 0
            console.log("MISSING DATA!", this);
            return
        }
        //rules for byr
        if (!byrRegex.test(this.byr)) {
            this.valid = 0
            console.log("INVALID BYR", this);
            return
        }
        // rules for iyr
        if (!iyrRegex.test(this.iyr)) {
            this.valid = 0
            console.log("INVALID IYR", this);
            return
        }
        // rules for eyr
        if (!eyrRegex.test(this.eyr)) {
            this.valid = 0
            console.log("INVALID EYR", this);
            return
        }
        // rules for hgt
        if (!hgtRegex.test(this.hgt)) {
            this.valid = 0
            console.log("INVALID HGT", this);
            return
        }
        // rules for hcl
        if (!hclRegex.test(this.hcl)) {
            this.valid = 0
            console.log("INVALID HCL", this);
            return
        }
        // rules for ecl
        if (!eclRegex.test(this.ecl)) {
            this.valid = 0
            console.log("INVALID ECL", this);
            return
        }
        // rules for pid
        if (!pidRegex.test(this.pid)) {
            this.valid = 0
            console.log("INVALID PID", this);
            return
        }
    }
}
```

* byr (Birth Year) - four digits; at least 1920 and at most 2002.
* iyr (Issue Year) - four digits; at least 2010 and at most 2020.
* eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
* hgt (Height) - a number followed by either cm or in:
** If cm, the number must be at least 150 and at most 193.
** If in, the number must be at least 59 and at most 76.
* hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
* ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
* pid (Passport ID) - a nine-digit number, including leading zeroes.
* cid (Country ID) - ignored, missing or not.
