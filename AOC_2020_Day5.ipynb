```javascript
const formatPartitions = (parts) => {
    const partArray = parts.split(",");
    const twoDPartArray = partArray.map(part => [...part]);
    return twoDPartArray
}
```

```javascript
const twoDPartitionArray = formatPartitions(entries);
```

```javascript
const findRow = (rowArr) => {
    return rowArr.reduce((acc, curr) => {
        const floor = curr === "F" ? acc[0] : acc[0] + (((acc[1]+1)-acc[0])/2);
        const ceiling = curr === "F" ? acc[0] + ((((acc[1]+1)-acc[0])/2)-1) : acc[1];
        if (floor === ceiling) {
            return floor
        }
        return [floor, ceiling]
    }, [0, 127]);
}
```

```javascript
const findColumn = (colArr) => {
    return colArr.reduce((acc, curr, idx) => {
        const floor = curr === "L" ? acc[0] : acc[0] + (((acc[1]+1)-acc[0])/2);
        const ceiling = curr === "L" ? acc[0] + ((((acc[1]+1)-acc[0])/2)-1) : acc[1];
        if (idx === 2) {
            return ceiling
        }
        return [floor, ceiling]
    }, [0, 7]);}
```

```javascript
const findSeatId = (row, col) => {
    return (row*8)+col
}
```

```javascript
const findSeatIds = (twoDPartArr) => {
    const seatIds = twoDPartArr.map((part) => {
        const partRow = part.slice(0, 7);
        const partCol = part.slice(7);
        const row = findRow(partRow);
        const col = findColumn(partCol);
        const seatId = findSeatId(row, col);
        return seatId
    });
    const sortedSeatIds = seatIds.sort((a,b) => a-b);
    return sortedSeatIds
}
```

```javascript
const seatIds = findSeatIds(twoDPartitionArray);
const highestSeatId = Math.max(...seatIds);
console.log(highestSeatId);
```

```javascript
const findMissingSeat = (seatIdArr) => {
    let currentSeat = Math.min(...seatIdArr);
    let missingSeat;
    for (let i = 0; i < seatIdArr.length; i++) {
        if (seatIdArr[i] !== currentSeat) {
            missingSeat = currentSeat
            break
        }
        currentSeat += 1
    }
    return missingSeat
}
```

```javascript
const missingSeat = findMissingSeat(seatIds);
console.log(missingSeat);
```
