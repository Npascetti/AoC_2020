```javascript
const formatFormGroupAnswers2 = (stringyInput) => {
    const splitByEmptyLine = stringyInput.split(/\n\s*\n/);
    const groupsArray = splitByEmptyLine.map((entry) => {
        return entry.split(/[ \n]/)
    });
    const collapsedGroupsArray = groupsArray.map((group) => {
        const flattendString = group.reduce((acc, curr) => {
            acc += curr
            return acc
        }, "");
        return [...flattendString]
    });
    const groupsObjectArray = collapsedGroupsArray.map((group, idx) => {
        const groupObject = group.reduce((acc, curr) => {
            if (!acc[curr]) {
                acc[curr] = 0
            }
            acc[curr] += 1
            return acc
        }, {});
        groupObject["validCount"] = groupsArray[idx].length
        return groupObject
    })
    return groupsObjectArray
}
```

```javascript
const formattedGroupAnswers2 = formatFormGroupAnswers2(entries);
```

```javascript
const getGroupsYesCount2 = (groupsObjArr) => {
    const groupsYesCount = groupsObjArr.reduce((acc, curr) => {
        for (const [key, val] of Object.entries(curr)) {
            if (key === "validCount") {
                continue;
            }
            if (val === curr.validCount) {
                acc += 1
            }
        }
        return acc
    }, 0)
    return groupsYesCount
}
```
