## Part 1

```javascript
const tobogganStringToArray = (tobStr) => {
    const tobArr = tobStr.split(",");
    const twoDTobArr = tobArr.map((row) => {
        return [...row]
    })
    return twoDTobArr
}
```

```javascript
const twoDTobogganArray = tobogganStringToArray(tobogganString);
```

```javascript
const findOpensAndTrees = (tobArr, over, down) => {
    let posArr = [];
    let overMove = over;
    for (let i = down; i < tobArr.length; i+=down) {
        overMove = overMove % 31;
        posArr = [...posArr, tobArr[i][overMove]]
        overMove += over;
    }
    return posArr
}
```

```javascript
const openTree31 = findOpensAndTrees(twoDTobogganArray, 3, 1)
const trees31 = openTree31.filter(openOrTree => openOrTree === "#");
console.log(trees31.length)
```

## Part 2
### The slopes to be tested
* first element is across, sescond element is down

```javascript
const slopes = [[1,1], [3,1], [5,1], [7,1], [1,2]];
```

```javascript
const findPartTwoProduct = (slopeArr) => {
    const treeCountArray = slopeArr.map((slope) => {
        const opensAndTrees = findOpensAndTrees(twoDTobogganArray, ...slope);
        return opensAndTrees.filter(openOrTree => openOrTree === "#").length;
    })
    return treeCountArray.reduce((acc, curr) => {
        return acc * curr
    }, 1)
}
```

```javascript
const part2Product = findPartTwoProduct(slopes);
console.log(part2Product)
```
